```python
### Langkah 1 :
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```python
### Langkah 2 :
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pad alist
for url in urls:
  image = io.imread(url)                            #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)   #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)    #convert color to RGB
  final_frame = cv.hconcat((image, image_2))        #concatenate image
  cv2_imshow(final_frame)
  print('\n')
```

```python
### Langkah 3 :
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi, " x ", lebar)
cv2_imshow(image_2)
```

```python
### Langkah 4 :
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

2. Buat garis vertikal dan garis menyilang diagonal pada image keluaran

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis vertikal ditengah image
for x in range (tinggi):
  image_3[x, int((lebar)/2)] = [255,255,255]

# membuat garis diagonal menyilang
for y in range (tinggi):
  image_4[y,int(y*(lebar/tinggi))] = [255,255,255]
  image_4[y,lebar-1-int(y*(lebar/tinggi))] = [255,255,255]

final_frame1 = cv.hconcat((image_2, image_3))
final_frame2 = cv.hconcat((image_2, image_4))
print('Garis Vertikal Ditengah Image')
cv2_imshow(final_frame1)
print('\n')
print('Garis Diagonal Menyilang ')
cv2_imshow(final_frame2)
```

3. Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang tertentu

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis vertikal ditengah image
for x in range ((tinggi - 10)//2, (tinggi - 100)//2 + 230 ):
  image_3[int((tinggi)/2 + 30), x] = [255,255,255]


final_frame = cv.hconcat((image_2, image_3))
print('Garis Horizontal ditengah dengan panjang tertentu ')
cv2_imshow(final_frame)
```

4. Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar


```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat kotak menggunakan kumpulan pixel warna putih
for x in range ((tinggi - 100)//2, (tinggi - 90)//2 + 100 ):
  for y in range((lebar + 20) //2, (lebar + 100) // 2 + 150):
    image_3[x+10,y-80] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
print('Kotak menggunakan kumpulan pixel warna putih')
cv2_imshow(final_frame)
```

Jawaban soal 2 :

Dilakukan conversi BGR2RGB dengan matplotlib adalah karena saat pertama di load format gambar yang digunakan oleh opencv adalah BGR bukan RGB yang merupakan standart format penampilan gambar pada kebanyakan layar

```python
## Jawaban soal 3

def draw_rectangle(image, xstart, ystart, xlength, ylength, color):
    for x in range(xlength):
        for y in range(ylength):
            image[ystart + y, xstart + x] = color

image_ktp_3 = image_ktp_2.copy()
draw_rectangle(image_ktp_3, 210, 2, 395, 30, [0, 0, 0])
draw_rectangle(image_ktp_3, 230, 33, 330, 30, [255, 255, 0])
plt.imshow(image_ktp_3)
plt.show()
```
